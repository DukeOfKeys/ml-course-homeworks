### заполним столбец с сертификатами самыми повторяющимися значениями 

```python
Certificate_mode = df['Certificate'].mode()[0]
print(f"самый популярный сертификат {Certificate_mode}") # рейтиг 18+
```

```python
df['Certificate'] = df['Certificate'].fillna(Certificate_mode)
```

### результат

```python
display(df.isnull().sum())
```

### внутри каждого жанра расчитываем медианное значение и заполняем этим проспуски 

```python
df['Runtime_of_Episodes'] = df['Runtime_of_Episodes'].astype(str).str.extract(r'(\d+)').astype(float)

df['Primary_Genre'] = df['Genre'].str.split(',').str[0].str.strip()

grouped_medians = df.groupby('Primary_Genre')['Runtime_of_Episodes'].transform('median')
df['Runtime_of_Episodes'] = df['Runtime_of_Episodes'].fillna(grouped_medians)

df['Runtime_of_Episodes'] = df['Runtime_of_Episodes'].fillna(df['Runtime_of_Episodes'].median())

df = df.drop(columns=['Primary_Genre'])

display(df.isnull().sum())
```

## метрики

```python
numeric_df = df.select_dtypes(include=['number'])

print("min, max, mean, median")
basic_stats = numeric_df.agg(['min', 'max', 'mean', 'median'])
display(basic_stats)

print("\n Mode")
mode_stats = numeric_df.mode().iloc[0] 
display(mode_stats)

print("\n Перцентили  Квантили")
quantiles = numeric_df.quantile([0.05, 0.25, 0.50, 0.75, 0.95])
display(quantiles)

print("\n Variance, Skewness, Kurtosis")
advanced_stats = pd.DataFrame({
    'Variance': numeric_df.var(),
    'Skewness': numeric_df.skew(),
    'Kurtosis': numeric_df.kurt()
})
display(advanced_stats)
```

```python
print("--- До Target Encoding ---")
display(df[['Series_Title', 'Genre', 'IMDB_Rating']].head(3))

# Считаем средний рейтинг для каждого жанра и заменяем текст на это число
target_mean = df.groupby('Genre')['IMDB_Rating'].mean()
df['Genre_Target_Encoded'] = df['Genre'].map(target_mean)

print("\n--- После Target Encoding ---")
display(df[['Series_Title', 'Genre', 'Genre_Target_Encoded', 'IMDB_Rating']].head(3))
```

```python
from category_encoders import HashingEncoder

# Указываем колонку и количество новых измерений (n_components), в которые "сожмем" актеров
hash_encoder = HashingEncoder(cols=['Star1'], n_components=8)

# Применяем к датасету
df_hashed = hash_encoder.fit_transform(df)

# Теперь вместо одной текстовой колонки Star1 у нас появятся колонки col_0, col_1 ... col_7 с числами
display(df_hashed.head())
```

```python
# Находим топ-5 самых частых сертификатов
top_5_certs = df['Certificate'].value_counts().nlargest(5).index

# Все, что не входит в топ-5, заменяем на 'Other'
# Метод .where() оставляет значение, если условие True, иначе меняет на указанное ('Other')
df['Certificate_Cleaned'] = df['Certificate'].where(df['Certificate'].isin(top_5_certs), 'Other')
```

```python
# Вытаскиваем первые 4 цифры из строки и переводим в числа
df['Start_Year'] = df['Runtime_of_Series'].astype(str).str.extract(r'^(\d{4})').astype(float)
```

### распределение рейтингов

```python
plt.figure(figsize=(10, 6))

# Строим гистограмму с наложением плавной кривой (KDE)
sns.histplot(data=df, x='IMDB_Rating', bins=30, kde=True, color='royalblue')

plt.title('Распределение рейтингов IMDb', fontsize=15, fontweight='bold')
plt.xlabel('Рейтинг (IMDB_Rating)', fontsize=12)
plt.ylabel('Количество сериалов', fontsize=12)
plt.show()
```

### Зависимость рейтинга от количества голосов  

```python
fig = px.scatter(
    df, 
    x='No_of_Votes', 
    y='IMDB_Rating', 
    hover_data=['Series_Title'],
    color='IMDB_Rating',        
    color_continuous_scale='Viridis',
    title='Зависимость рейтинга от количества голосов ',
    opacity=0.7
)

# Делаем фон чистым
fig.update_layout(template='plotly_white')
fig.show()
```

```python
plt.figure(figsize=(12, 6))

top_certs = df['Certificate'].value_counts().nlargest(6).index
df_filtered = df[df['Certificate'].isin(top_certs)]

sns.violinplot(
    data=df_filtered, 
    x='Certificate', 
    y='Runtime_of_Episodes', 
    color='lightgray', 
    inner=None 
)

sns.boxplot(
    data=df_filtered, 
    x='Certificate', 
    y='Runtime_of_Episodes', 
    palette='Set2', 
    width=0.3
)

plt.title('Разброс длительности серий по возрастным рейтингам', fontsize=15)
plt.xlabel('Возрастной рейтинг (Certificate)', fontsize=12)
plt.ylabel('Длительность (минуты)', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

df['Main_Genre'] = df['Genre'].str.split(',').str[0].str.strip()
top_genres = df['Main_Genre'].value_counts().head(10)

sns.barplot(
    x=top_genres.values, 
    y=top_genres.index, 
    hue=top_genres.index,  
    palette='magma', 
    legend=False            
)

plt.title('Топ-10 самых популярных жанров', fontsize=15)
plt.xlabel('Количество сериалов', fontsize=12)
plt.ylabel('Главный жанр', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))

numeric_df = df.select_dtypes(include=['number'])

corr_matrix = numeric_df.corr()

# Рисуем тепловую карту
sns.heatmap(
    corr_matrix, 
    annot=True,        
    cmap='coolwarm',   
    fmt='.2f',           
    linewidths=1,       
    vmin=-1, vmax=1    
)

plt.title('Тепловая карта корреляций числовых признаков', fontsize=15)
plt.show()
```
